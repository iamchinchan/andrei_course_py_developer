```python
original = [1, 2, [3, 4]]

copy_one = original.copy()
copy_two = original[::]

# Modifying a nested list affects all copies
original[2][0] = "changed"

print(copy_one)  # Output: [1, 2, ['changed', 4]]
print(copy_two)  # Output: [1, 2, ['changed', 4]]
```

```python
import copy

# A nested list (2D matrix)
original = [[1, 2], [3, 4]]

# Perform a deep copy
deep_copied_list = copy.deepcopy(original)

# Modify the nested element in the original list
original[0][0] = 999

# Verify independence
print("Original:", original)  # Output: [[999, 2], [3, 4]]
print("Deep Copied:", deep_copied_list)  # Output: [[1, 2], [3, 4]] -> Unchanged!
```

In Python, there is no functional difference between list.copy() and list[::]. Both methods perform a shallow copy of the list.However, they differ slightly in syntax, intent, and performance

New Container, Same References: Both create a new list object in memory.Shallow Copy Limitation: If your list contains nested objects (like lists or dictionaries), both methods copy only the references to those sub-objects. Modifying a nested object changes it in both lists.

 Performance (Micro-Optimization)list[::] is slightly faster: Slicing is executed directly at the C-level bytecode layer in Python. It does not look up an attribute name.list.copy() is slightly slower: It requires a Python attribute lookup (finding the .copy method on the object) before executing.
