```python
x = 100


def outer(y=x):
    x = 20

    def inner(z):
        nonlocal x
        x += z
        return x

    return inner


f = outer()
# Tricky Walrus + Scope
while (val := f(5)) < 30:
    print(f"Val in loop: {val}")
    x = 500  # Does this affect the 'x' inside outer?

print(f"Final x: {x}")
print(f"Final f(0): {f(0)}")
```
