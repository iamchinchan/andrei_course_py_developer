To understand exactly how the address stays the same while the memory expands, you have to split a Python list into two separate physical pieces: the Outer Wrapper and the Inner Array.

Here is the exact step-by-step breakdown of how Python pulls off this magic trick.

1. The Setup: Variables and Pointers
The Variable (my_list): This is just a name tag. It doesn't hold data. It just points to an address.

The Outer Wrapper (The Sticky Note): This is the main object at a fixed memory address (e.g., Address 1000). When you call id(my_list), it returns 1000.

The Inner Array (The Filing Cabinet): This is a separate block of memory somewhere else (e.g., Address 8000). It contains the actual slots holding pointers to your data.

Phase 1: Creating the List

```python
my_list = [10, 20]
```

Python creates the Outer Wrapper at Address 1000. It writes three things on this sticky note:

Current Size: 2 items

Max Capacity: 4 slots

Target Address: Go to Address 8000 to find the slots.

Python creates the Inner Array at Address 8000 with 4 slots.

Slot 1 points to the number 10.

Slot 2 points to the number 20.

Slots 3 and 4 are completely empty.

Phase 2: Shrinking (No memory moves)

```python
my_list.pop()
```

Python goes to the Outer Wrapper at 1000 and changes "Current Size" from 2 to 1.

It goes to the Inner Array at 8000 and deletes the pointer in Slot 2.

Result: The Outer Wrapper is still at 1000. The Inner Array is still at 8000. Nothing moved.

Phase 3: Expanding (The Magic Trick)
Now we add three more items, pushing the total to 4 items. The Inner Array at 8000 is now 100% full.

```python
my_list.append(50)  # This breaks the capacity!
```

The Inner Array at 8000 has no more physical space to grow. Here is exactly how Python handles it without changing the list's id():

Allocate New Space: Python goes to a completely new area of RAM, say Address 9000, and builds a brand-new, bigger Inner Array with 8 slots.

Move the Data: It copies the 4 pointers from the old array at 8000 into the new array at 9000. Then it puts 50 into the 5th slot.

Destroy the Old: It deletes the old array at 8000.

The Update (The Secret): Python goes back to the Outer Wrapper at Address 1000. It takes an eraser, rubs out "Target Address: 8000", and writes "Target Address: 9000".

The Conclusion
When you run id(my_list) after the expansion, Python looks at the Outer Wrapper. The Outer Wrapper never moved. It has been sitting at Address 1000 the entire time.

All Python did was change the coordinates written on the sticky note. That is how the list's memory footprint can wildly expand and move around inside the RAM, but the id() of your list remains mathematically identical forever.
