```python
prices = [10, 20, 30, 45, 50, 10]

# We want to process prices until we hit one over 40
# The walrus operator (p := ...) saves the current price for printing
# while simultaneously letting the while loop check the condition.

index = 0
while index < len(prices) and (p := prices[index]) <= 40:
    print(f"Processing item with price: {p}")
    index += 1

print(f"Stopped at price: {p}")  # 'p' is still available here!
```

In almost every other programming language, a variable defined inside a while loop condition would die as soon as the loop ends.

But Python is different. Python does not have "block scope" (the way languages like C++ or Java do). In Python, if you define a variable, it sticks around until the function or module it is in finishes.

The Scope Reality
When you write while (p := prices[index]) <= 40:, the Walrus operator assigns the value to p in the local scope of the function you are currently in.

Inside the loop: p is alive and holds the value of the current item being processed.

After the loop: p is still alive and holds the last value it was assigned before the loop condition failed.

Why this is a "Feature," not a bug
This allows you to do "Post-Mortem Analysis" on your loops.

The "Danger Zone"
Because Python handles scope this way, you have to be careful. If you try to use p before the loop has ever run, Python will raise a NameError because p hasn't been created yet.

Think of it this way:

The Walrus operator is like leaving a trail of breadcrumbs in the function.

Once you create p with the Walrus, that variable exists for the rest of that function's life.
