```python
globar_var = 10


def outer():
    nonlocal_var = 20

    def inner():
        global globar_var
        nonlocal nonlocal_var
        globar_var += 5
        nonlocal_var += 5
        print(globar_var)
        print(nonlocal_var)

    inner()


outer()
print(globar_var)
# print(nonlocal_var)  # This will raise an error because 'nonlocal_var' is not defined in the global scope
```

```python
var = 10


def outer():
    var = 20

    def inner():
        nonlocal var
        var += 5
        print(var)

    print(f"inside outer before calling inner fn: var is  : {var}")
    inner()
    print(f"inside outer after calling inner fn: var is  : {var}")


outer()
print(f"global var is : {var}")
```

```python
var = 10

print(f"global var is : {var}")


def outer():
    var = 20

    def inner():
        global var
        var += 5
        print(var)

    print(f"inside outer before calling inner fn: var is  : {var}")
    inner()
    print(f"inside outer after calling inner fn: var is  : {var}")


outer()
print(f"global var after outer fn call is : {var}")
```
