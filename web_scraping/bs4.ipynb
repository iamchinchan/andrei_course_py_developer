```python
import requests
import re
from bs4 import BeautifulSoup as bs
```

```python
url = "https://news.ycombinator.com/?p=1"
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}
res = requests.get(url=url, headers=headers)
```

```python
print(res)
print(type(res))
print(res.text)
print(type(res.text))
```

```python
soup = bs(res.text, "lxml")
print(soup)
print(type(soup))
print(soup.title)
print(type(soup.title))
print(soup.body.contents)
print(type(soup.body.contents))
print(type(soup.body.contents[0]))
print(soup.find("a"))
```

```python
ala = soup.select("a")
print(type(ala))
print(ala)
```

```python
news = soup.select(".titleline")
subtexts = soup.select(".subtext")
print(news)
print("--------------------")
print(subtexts)
if len(news) == len(subtexts):
    pass
else:
    print("Mismatch!!")
```
