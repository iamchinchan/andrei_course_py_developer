```python
def pick_random_page_index(already_compared_index):
    while True:
      random_index = random.randint(0,len(data)-1)
      if random_index not in already_compared_index:
        already_compared_index.add(random_index)
        return random_index
      
```

```python
def get_winner(first_player_index:int, second_player_index:int):
  if data[first_player_index]["follower_count"]==data[second_player_index]["follower_count"]:
    return 'd'
  #d means draw
  elif data[first_player_index]["follower_count"]>data[second_player_index]["follower_count"]:
    return 'a'
  else:
    return 'b'
```

```python
game_over =False
score=0
already_compared_index = set()
first_player_index=-1
print(logo)
while not game_over and len(already_compared_index)!=len(data):
  if first_player_index==-1:
    first_player_index = pick_random_page_index(already_compared_index)
  second_player_index = pick_random_page_index(already_compared_index)
  won=False
  print(f"Compare A: {data[first_player_index]["name"]}, {data[first_player_index]["description"]}, from {data[first_player_index]["country"]}.")

  print(f"{vs}")

  print(f"Against B: {data[second_player_index]["name"]}, {data[second_player_index]["description"]}, from {data[second_player_index]["country"]}.")

  result = get_winner(first_player_index,second_player_index)

  input_from_user = input("Who has more followers? Type 'A' or 'B'").lower()

  if result=='d':
    # it is a draw: so user can pick anything and will always win
    won=True
  else:
    # result is eiother a or b
    # it is not a tie
    if result == input_from_user:
      won = True

  if won:
    # if this round is won by the user
    score+=1
    first_player_index = second_player_index
    print(f"Your guess was right!, your score is  {score}")
  else:
    print(f"Wrong guess!: correct answer was : {result}, Your final score is {score}")
    game_over = True


if len(already_compared_index)!=len(data):
  print(f"You have completed the whole game ! Congratulations: you got full score that can be achieved: {len(data)-1}")
```
