```python
with open("test.txt", "r") as f:
    print(f)
    print(f.readlines())
    f.seek(0)
    for line in f.readlines():
        print(line, end="")
    f.seek(0)
    print()
    print(f.read())
    f.seek(0)
    print(f.readline(), end="")
    print(f.readline(), end="")
    print(f.readline(), end="")
```
