```python
def counter(func):
    count = 0

    def inner(*args, **kwargs):
        nonlocal count
        count += 1
        print(f"Function called {count} times")
        return func(*args, **kwargs)

    return inner
```

```python
def printer(name="Jatin"):
    print(f"Hello, {name}!")


printer = counter(printer)
```

```python
@counter
def printer(name="Jatin"):
    print(f"Hello, {name}!")
```
