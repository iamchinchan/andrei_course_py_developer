```python
def counter(func):
    count = 0

    def inner(*args, **kwargs):
        nonlocal count
        count += 1
        print(f"Function called {count} times")
        return func(*args, **kwargs)

    return inner
```

```python
def printer(name="Jatin"):
    print(f"Hello, {name}!")


printer = counter(printer)
```

```python
@counter
def printer(name="Jatin"):
    print(f"Hello, {name}!")
```

```python
from time import time


def performance(fn):
    def wrapper(*args, **kwargs):
        t1 = time()
        result = fn(*args, **kwargs)
        t2 = time()
        print(f"It took {t2-t1} seconds to run")
        return result

    return wrapper
```

```python
@performance
def dummy():
    for i in range(100000000):
        pass
```
