```python
def outer_immutable():
    x = 10  # Immutable integer

    def inner():
        # READING works perfectly
        print("Reading x:", x)

    inner()
```

```python
def outer_immutable():
    x = 10  # Immutable integer

    def inner():
        x = 50  # This creates a new local variable 'x' in the inner function's scope, shadowing the outer 'x'.
        print("Reading x:", x)

    inner()
```

```python
outer_immutable()
```

```python
def outer_crash():
    x = 10

    def inner():
        # Python sees an assignment (=) and assumes 'x' is a brand new LOCAL variable.
        x = x + 1  # CRASH: UnboundLocalError! (Local 'x' referenced before assignment)
        return x

    inner()
```

```python
outer_crash()
```

```python
x = 10  # Global variable


def my_function():
    print(x)  # Throws UnboundLocalError!
    x = 5  # Because of this assignment, 'x' is local for the WHOLE function
```

```python
my_function()
```
