```python
def super_func(a, b, *args, c=10, **kwargs):
    print(a)
    print(b)
    print(args)
    print(c)
    print(kwargs)


super_func(1, 2, 3, 4, 5, c=20, d=30, e=40)
```

```python
# This works because functions expect key=value arguments
def my_func(three, six, seven):
    print(three, six)


my_func(**{"three": 3, "six": 6, "seven": 7})
```

```python
# This works because you are unpacking it into a new set of brackets
test = {**{"three": 3, "six": 6}, "eight": 8}
```
