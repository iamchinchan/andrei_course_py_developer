```python
from collections import namedtuple

# 1. Define the blueprint
Person = namedtuple("Persona", ["name", "age", "city"])

# 2. Create an instance (it acts like a regular object)
user = Person("Jatin", 26, "India")

# 3. Access by name (much more readable than user[0])
print(user.name)  # Output: Jatin
print(user.age)  # Output: 26
```

```python
print(type(user))
# persona nor person, person is just a variable name, persona is the internal name of the class
```

```python
print(type(Person))
```

```python
# 1. Define the __init__ method that saves the values
def __init__(self, x, y):
    self._x = x
    self._y = y


# 2. Define the 'getters' so you can use point.x instead of point._x
# We use the @property decorator to make them behave like attributes
@property
def x(self):
    return self._x


@property
def y(self):
    return self._y


# 3. Use the 'type' factory to build the class
# We inherit from 'tuple' so it acts like a sequence if needed
Point = type("Point", (tuple,), {"__init__": __init__, "x": x, "y": y})

# --- Test the manual implementation ---
p = Point(10, 20)
print(f"X: {p.x}, Y: {p.y}")
```

```python
# The only way to save data in a tuple-subclass is via __new__
def my_new(cls, x, y):
    # This is the "Concrete" step. Memory is locked here.
    return tuple.__new__(cls, (x, y))


# The property "Trapdoors"
@property
def x(self):
    return self[0]  # Fetch from the tuple index


@property
def y(self):
    return self[1]


# Build the class correctly
Point = type("Point", (tuple,), {"__new__": my_new, "x": x, "y": y})

# Now it works!
p = Point(10, 20)
print(f"X: {p.x}, Y: {p.y}")
```
