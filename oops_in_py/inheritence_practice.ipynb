```python
# 1. Go to the class (agent.__class__)
# 2. Ask the class for its parents (.__bases__)

print(agent.__class__.__bases__)
# Output: (<class '__main__.NetworkNode'>, <class '__main__.UserInterface'>)
```

1. In Python: __class__

For an instance (agent.__class__), it points to its Blueprint (e.g., <class 'JarvisAgent'>).

For a Class (JarvisAgent.__class__), it points to its Blueprint, which is the metaclass (e.g., <class 'type'>).

2. In the C-Engine: ob_type

Because both instances and classes are PyObject structs at their core, they both possess the ob_type pointer.

When you call the built-in type(obj) function in Python, the engine simply drops into C, reads the ob_type pointer of that specific object, and hands it back to you.

Here is the exact summary of what you just cracked:

__class__ = The Maker (Who built me?)

Instance's maker is the Class. (agent.__class__ -> JarvisAgent)

Class's maker is type. (JarvisAgent.__class__ -> type)

__base__ / __bases__ = The Parents (Who did I inherit from?)

Used strictly for sharing code and MRO. (JarvisAgent.__base__ -> NetworkNode)

The Golden Rule: Instances only know their Maker (__class__). Classes know both their Maker (__class__) and their Parents (__base__ / __bases__).
