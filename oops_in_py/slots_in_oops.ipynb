__slots__ is a Python feature used strictly for performance and memory optimization. It physically changes how the C-engine builds your object.

```python
class Dog:
    # Tells Python: ONLY these two variables will ever exist.
    __slots__ = ["name", "age"]

    def __init__(self, name, age):
        self.name = name
        self.age = age
```

2. What happens in the C-Engine
Normally, Python objects use a __dict__ (a heavy, dynamic hash table) to store variables.

When you declare __slots__:

Python refuses to create the __dict__.

Instead, it bakes exact, fixed memory slots directly into the object's C-struct.

3. The Pros (Why use it?)
Massive RAM Savings: A standard object with a __dict__ takes about ~150-300 bytes of memory. A slotted object takes about ~40-50 bytes. If you are generating millions of data points, this saves Gigabytes of RAM.

Faster Access: Looking up dog.name skips the dictionary hash-search and jumps straight to a hardcoded C-level memory index (just like accessing tuple[0]).

4. The Catch (The Trade-off)
You completely lose dynamic flexibility. Because there is no dictionary, you cannot add new variables on the fly.

```python
d = Dog("Rex", 5)
d.breed = "Husky"  # CRASH! AttributeError: 'Dog' object has no attribute 'breed'
```

```python
print(d.name)
```

```python
d.name = "Max"
print(d.name)
# it can be chnaged
```

Summary: __slots__ gives you the lightning-fast, tightly packed memory structure of a tuple, but leaves those specific slots "open" so __init__ can write to them. Use it only when you have a massive amount of identical objects and need to save memory.
