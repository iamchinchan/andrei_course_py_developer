```python
class Base:
    def setup(self):
        print("Base setup")


class ParentA(Base):
    def setup(self):
        print("ParentA setup")
        super().setup()  # The baton pass!


class ParentB(Base):
    def setup(self):
        print("ParentB setup")
        super().setup()  # The baton pass!


class Child(ParentA, ParentB):
    def setup(self):
        print("Child setup")
        super().setup()  # The baton pass!
```

```python
c = Child()
```

```python
print(c)
```

```python
c.setup()
```

1. What does "Temporary Proxy Object" mean?
Think of super() as a disposable remote control.
When you type super(), Python quickly builds a remote control, points it at the Parent class, presses the button you asked for, and then immediately throws the remote in the trash. It just exists for one second to trigger the Parent.

2. The 4 Exact Scenarios You Asked About
Here is exactly what happens with parameters, calling, and not calling super().

Case A: The Child has NO __init__ at all.
What happens: Yes, the Parent is called automatically!

Why: If you don't write an __init__ in the Child, Python just uses the Parent's __init__ instead. You don't even need to type super().

Case B: The Child HAS an __init__, but you FORGET to call super().
What happens: The Parent is NOT called. It is completely ignored.

The Danger: If the Parent was supposed to set up something important (like self.database_connection), it never happens. When your Child tries to use the database later, the app will crash because the variable doesn't exist.

Case C: The Parent requires a parameter, but you DON'T pass it.

```python
class Parent:
    def __init__(self, name):  # Parent demands a name
        self.name = name


class Child(Parent):
    def __init__(self, name, age):
        super().__init__()  # FORGOT TO PASS NAME!
```


What happens: Instant Crash.

Error: Python throws a TypeError: __init__() missing 1 required positional argument: 'name'.

Case D: How parameters SHOULD be passed.
The Rule: You must pass exactly what the Parent asks for inside the super() parentheses.

```python
class Child(Parent):
    def __init__(self, name, age):
        super().__init__(name)  # Pass the name to the Parent
        self.age = age  # Child handles the age itself
```

Where MUST it be called?
If you are overriding a method (like __init__), you usually call super() on the very first line inside that method. You let the Parent do its basic setup first, and then you add the Child's custom stuff underneath it.
