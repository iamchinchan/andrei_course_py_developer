```python
class User:
    def sing(self):
        return "la la la"

    def attack(self):
        return "does a normal attack!"


class Wizard(User):
    def __init__(self, name):
        self.name = name

    def cast_spell(self):
        return "abracadabra"

    def attack(self):
        print(User.attack(self))  # call the parent method
        return f"{self.name} does a spell attack!"


class Archer(User):
    def __init__(self, name):
        self.name = name

    def attack(self):
        return f"{self.name} does a bow attack!"


def player_attack(player):
    return player.attack()


wizard1 = Wizard("Merlin")
archer1 = Archer("Robin")
print(dir(wizard1))
for player in [wizard1, archer1]:
    print(player_attack(player))
```
