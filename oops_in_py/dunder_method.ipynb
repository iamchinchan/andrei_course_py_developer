```python
class Toy:
    def __init__(self, color, age):
        self.color = color
        self.age = age
        self.my_dict = {"brand": "Hasbro", "type": "Car"}

    def __str__(self):
        return f"{self.color}"

    def __len__(self):
        return self.age

    def __call__(self):
        return "yesss?"

    def __del__(self):
        print("deleted!")

    def __getitem__(self, i):
        return self.my_dict[i]
```

```python
action_figure = Toy("red", 10)
print(action_figure)
```

```python
print(action_figure.__str__())
print(str(action_figure))
```
