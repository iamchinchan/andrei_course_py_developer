Here is the absolute, bottom-line truth about Dunder (Double UNDERscore) methods: They are just hooks.

As you saw with the C-engine, Python has built-in behaviors (like printing, adding, or checking length) written in C. Dunder methods are your way to wire your custom classes into those hardcoded C-behaviors so your objects act like native Python data types.

Here is the step-by-step breakdown of the ones you actually need to know, grouped by what they hook into.

1. The Lifecycle Hooks (Birth and Setup)
These control how your object is created in memory and initialized.

__new__(cls): The actual memory allocator. This is what asks the OS for RAM to build the "box." You rarely touch this unless you are building strict singletons.

__init__(self): The setup logic. It runs immediately after the box is built to put your data inside.

```python
class Agent:
    def __init__(self, name):
        self.name = name  # Wires the data into the instance
```

2. The Representation Hooks (What we just debugged)
These control what the interpreter returns when you inspect an object.

__str__(self): Hooks into print() and str(). Meant to be clean and human-readable.

__repr__(self): Hooks into the developer console. Meant to show the exact code needed to recreate the object (useful for debugging your agent's state).

```python
class Tool:
    def __str__(self):
        return "Web Scraper Tool"

    def __repr__(self):
        return "<Tool(type='scraper', active=True)>"
```

3. The Math and Logic Hooks (Operator Overloading)
These intercept standard math symbols and comparison operators so you don't have to write clunky methods like a.add_to(b).

__add__(self, other): Hooks into the + symbol.

__eq__(self, other): Hooks into the == symbol.

```python
class Vector:
    def __init__(self, x):
        self.x = x

    def __add__(self, other):
        return Vector(self.x + other.x)  # Now vec1 + vec2 actually works
```

4. The Container Hooks (Making objects act like lists/dicts)
These allow your custom classes to be iterated over, sliced, or measured.

__len__(self): Hooks into len().

__getitem__(self, key): Hooks into bracket notation obj[key].

```python
class MemoryStore:
    def __init__(self):
        self.history = ["Prompt 1", "Prompt 2"]

    def __len__(self):
        return len(self.history)  # Now len(my_store) returns 2
```

5. The Execution Hook (The absolute best one for AI)
__call__(self, *args): This allows you to execute an instance exactly as if it were a function. When building automated tools, this keeps your code incredibly clean.

```python
class Summarizer:
    def __call__(self, text):
        return f"Summarizing: {text}"


# You build the instance once
summarize_tool = Summarizer()

# Then you just 'call' the object directly
result = summarize_tool("Heavy systems data...")
```

action_figure.__str__ is a bound method because you explicitly wrote def __str__(self): inside your Toy class. It is standard Python code.

action_figure.__repr__ is a method-wrapper because you did not write one. The interpreter found nothing in Toy, so it fell back to the raw C-code inside the object base class.

What __repr__ actually is:

While __str__ and __repr__ both return strings, they have completely different target audiences.

1. __str__ is for the User (Display)
It should be clean, pretty, and readable.

Output: "Iron Man Action Figure"

2. __repr__ is for the Developer (Debugging)
It stands for "Representation." Its strict goal is to be unambiguous. The golden rule of Python is that __repr__ should return the exact string of Python code required to perfectly recreate that object.

Output: "Toy(name='Iron Man Action Figure', price=20.00)"

The Fallback Rule:
If you type an object's name in the terminal and hit Enter, the console calls __repr__. If you use print(object), it calls __str__.

However, if you do not define __str__, Python will automatically fall back to calling __repr__ when printing so the program doesn't crash.
