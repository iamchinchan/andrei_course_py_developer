```python
from pypdf import PdfReader, PdfWriter
```

```python
with PdfReader("merged.pdf") as reader:
    for page in reader.pages:
        print(page.extract_text())
```

```python
with PdfReader("wtr.pdf") as wtr_reader:
    wtr_page = wtr_reader.get_page(0)
    with PdfWriter() as writer:
        with PdfReader("merged.pdf") as m_reader:
            for page in m_reader.pages:
                page.merge_page(wtr_page)
                writer.add_page(page)
        writer.write("watermarked.pdf")
```
