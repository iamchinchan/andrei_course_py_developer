```python
import pypdf
from pathlib import Path
```

```python
current_folder = Path.cwd()
print(current_folder)
print(type(current_folder))
```

```python
# Using iterdir() and manually checking for the .pdf extension
pdf_names = [
    file.name
    for file in current_folder.iterdir()
    if file.is_file() and file.suffix.lower() == ".pdf"
]

print(pdf_names)
```

```python
# Assuming current_folder is already defined
pdf_names = [file.name for file in current_folder.glob("*.pdf") if file.is_file()]

print(pdf_names)
```

```python
def pdf_combiner(pdf_list):
    with pypdf.PdfWriter() as writer:
        for pdf in pdf_list:
            writer.append(pdf)
        writer.write("merged.pdf")
```

```python
pdf_combiner(pdf_names)
```

```python
with pypdf.PdfReader("merged.pdf") as reader:
    print(reader)
    print(reader.get_num_pages())
```
