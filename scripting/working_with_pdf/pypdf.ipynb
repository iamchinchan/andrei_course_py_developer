```python
import pypdf
```

```python
with open("dummy.pdf") as f:
    print(f"f is {f}")
```

```python
with open("dummy.pdf", "rb") as f:
    print(f"f is {f}")
```

```python
with open("dummy.pdf", "rb") as f:
    print(f"f is {f}")
    reader = pypdf.PdfReader(f)
    print(f"reader is  : {reader}")
    print(reader.get_num_pages)
    print(reader.get_num_pages())
    page_obj = reader.get_page(0)
    print(f"type of page_obj is: {type(page_obj)}")
    print(f"page_obj : {page_obj}")
```

```python
with open("dummy.pdf", mode="rb") as f:
    reader = pypdf.PdfReader(f)
    page = reader.get_page(0)
    page.rotate(90)
    writer = pypdf.PdfWriter()
    writer.add_page(page)
    with open("tiltedDummy.pdf", "wb") as nf:
        writer.write(nf)
```
