```python
import time


def limit_handler(follower_bundle):
    try:
        while True and len(follower_bundle.followers) > 0:
            yield follower_bundle.followers.pop()
    except NetworkError as error:
        time.sleep(1000)


response = client.get_followers(did_fox)
for follower in limit_handler(response):
    if follower.display_name == "Deploio":
        print(f"following Deploio")
        client.follow(follower.did)
```

```python
response = client.get_followers(did_fox)
while len(response.followers) > 0:
    try:
        follower = response.followers.pop()
        if follower.display_name == "Deploio":
            print(f"following Deploio")
            client.follow(follower.did)
            break
    except NetworkError:
        time.sleep(1000)
```

```python
# 4 Like Bot
from atproto_client.exceptions import NetworkError

login = client.login(os.getenv("BLUESKY_HANDLE"), os.getenv("BLUESKY_PASSWORD"))
feed = client.get_timeline(limit=5).feed
print(len(feed))
for post_view in feed:
    print(post_view.post.record.text)
search_string = "Good morning!"
numberOfPosts = 2
count = 0
try:
    for item in feed:
        if item.post.record.text.find(search_string) > -1:
            count += 1
            client.like(item.post.uri, item.post.cid)
            print("Liked a post")
            if count >= numberOfPosts:
                break
except NetworkError as e:
    print(e.response)
```
