```python
def fib(n):
    a = 0
    b = 1
    for i in range(n):
        yield a
        b = a + b
        a = b - a
```

```python
for x in fib(20):
    print(x, end=" ")
```

```python
f = fib(5)
print(f)
print(list(f))
print(list(f))  # exhausts
```

```python
for i, x in enumerate(fib(5)):
    print(f"f{i+1}:{x}", end=",")
```
