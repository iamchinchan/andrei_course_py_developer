```python
class itr:
    def __init__(self, start, stop):
        self.start = start
        self.stop = stop
        self.index = start - 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.index + 1 < self.stop:
            self.index += 1
            return self.index
        else:
            raise StopIteration
```

```python
class dr:
    def __init__(self, start, stop):
        # start including and stop not including
        self.start = start
        self.stop = stop

    def __str__(self):
        return f"range({self.start},{self.stop})"

    def __getitem__(self, key):
        if (result := self.start + key) < self.stop:
            return result
        else:
            raise IndexError

    def __iter__(self):
        return itr(self.start, self.stop)
```

```python
r = dr(5, 10)
```
