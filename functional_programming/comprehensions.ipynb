```python
# dictioonary comprehension
sd = {"a": 1, "b": 2}
nd = {k: v**2 for k, v in sd.items()}
print(nd)
```

```python
# dictioonary comprehension
sd = {"a": 1, "b": 2}
nd = {k: v**2 for k, v in sd.items() if not v % 2}
print(nd)
```
